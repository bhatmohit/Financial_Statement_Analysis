# Data Preparation

## Calculate Accruals and Cash Flows (scaled by total assets)

Accruals = Net Income(ni) - Cash Flows from Net Operating Activities(oancf)


```python
df['accruals'] = (df['ni'] - df['oancf'])/df['at']
df['scaled_oancf'] = df['oancf']/df['at']
df[['accruals', 'scaled_oancf']]
```

## Data Cleaning

Remove inf values

```python
df[['accruals', 'scaled_oancf']].describe()
```

```python
df[['accruals', 'scaled_oancf']] = df[['accruals', 'scaled_oancf']].replace([-np.inf, np.inf], np.nan)
# df[['scaled_oancf']].replace([-np.inf, np.inf], np.nan)
df.dropna(subset=['accruals', 'scaled_oancf'])
df[['accruals', 'scaled_oancf']].describe().round(2)
```

# Future Returns Analysis

Explore how firms perform over following year based on current cash flows and accruals ranking

```python
# assign each observation into one of 5 groups based on accruals.
df['accruals_ranking'] = pd.qcut(df['accruals'], 5, labels=[1,2,3,4,5])

# do the same for cash flows
df['oancf_ranking'] = pd.qcut(df['scaled_oancf'], 5, labels=[1,2,3,4,5])
```

Examine future returns for firms with high vs low accruals

```python
accruals_summary = df.pivot_table(index=['accruals_ranking'], values=['annret'], aggfunc='median')

accruals_summary
```

Firms with high accruals (rank 5) perform worse than firms with low accruals (rank 1)

Examine future returns for high vs low cash flows

```python
cash_flow_summary = df.pivot_table(index=['oancf_ranking'], values=['annret'], aggfunc='median')
cash_flow_summary
```

Combine both

```python
summary = df.pivot_table(index=['accruals_ranking', 'oancf_ranking'], values=['annret'], aggfunc='median')
summary = summary.unstack('oancf_ranking')
summary
```

# Results

The above analysis concludes that: <br/>
Firms with high cash flows and low accruals perform better than firms with high accruals and low cash flows
